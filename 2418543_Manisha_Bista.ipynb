```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers
outlier_countries = detect_outliers_iqr(df,'score')
outlier_countries
```

```python
plt.scatter(df_South_asian_countries['Log GDP per capita'],df_South_asian_countries['score'],label="Score")
plt.scatter(outlier_countries['Log GDP per capita'],outlier_countries['score'],color='red',label="Outliers")
plt.xlabel('GDP per capita')
plt.ylabel('Happiness Score')
plt.title('GDP per capita vs Happiness Score')
plt.legend()
```

```python
social_support_correlation = df_South_asian_countries['Social support'].corr(df_South_asian_countries['score'],method='pearson')
corruption_perception_correlation = df_South_asian_countries['Perceptions of corruption'].corr(df_South_asian_countries['score'],method='pearson')
print("Pearson Correlation with score for South Asian Countries")
print("Social Support:",social_support_correlation)
print("Perceptions of corruption:",corruption_perception_correlation)
```

```python
def metrics_mean(DF):
  metrics = [DF['Log GDP per capita'].mean(), DF['Social support'].mean(), DF['Healthy life expectancy'].mean()]
  return metrics
metrics_SA = metrics_mean(df_South_asian_countries)
metrics_ME = metrics_mean(df_middle_east)
x = np.arange(3)
width = 0.35
plt.bar(x-width/2,metrics_SA,width,label='South Asian Countries')
plt.bar(x+width/2,metrics_ME,width,label='Middle East Countries')
plt.xlabel('Metrics')
plt.ylabel('Mean')
plt.title('Mean of Metrics Comparison')
plt.xticks(x,['Log GDP per capita','Social support','Healthy life expectancy'])
plt.legend()
```
